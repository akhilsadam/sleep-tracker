```python
import matplotlib.pyplot as plt
import numpy as np
class machine():
    pi = 3.14159265358    
    def __init__(self):
        self.kD = 2;
        self.kC = 10;
        self.THLD = 1;
        self.state = 1;
        self.debt = 0;
        self.cir = 0;
        self.sleep = 0;
        self.time = 0;
    def debtStep(self,stepLen):
        self.debt+=self.kD*stepLen*self.state
    def cirStep(self,stepLen):
        self.time +=stepLen
        self.cir=self.kC*np.sin(self.time*self.pi)
    def sleepStep(self):
        self.sleep = self.debt+self.cir
        if(self.sleep<self.THLD):
            self.state = -1
        else:
            self.state = 1
    def step(self,stepLen):
        self.debtStep(stepLen)
        self.cirStep(stepLen)
        self.sleepStep()
    def gTime(self):
        return self.time
    def gSleep(self):
        return self.sleep
    def gDebt(self):
        return self.debt
    def gCir(self):
        return self.cir
    def gState(self):
        return self.state
```

```python


time = []
sleep = []
debt = []
cir = []
state = []
Ntime = 100
sl = 40
Nstep = sl*Ntime

m = machine()

for i in range (0,Nstep):
    m.step(1/sl)
    time.append(m.gTime())
    sleep.append(m.gSleep())
    cir.append(m.gCir())
    debt.append(m.gDebt())
    state.append(m.gState())
plt.plot(time,sleep)
plt.plot(time,debt)
plt.plot(time,cir)
plt.plot(time,state)
```
