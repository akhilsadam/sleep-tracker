```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import numpy as np
import scipy.optimize as opt
import scipy.ndimage as ndi
import cv2
from scipy.signal import convolve2d as c2d
from datetime import datetime as dtime
from datetime import timedelta as td
#----------------------------------------------------------
def DT(dtr):
    dt0 = dtime.strptime(dtr, '%H%M');
    return (dt0.hour*60+dt0.minute);
day0 = "09-16";
dd0 = dtime.strptime(day0, '%m-%d');
def DD(n):
    dt0 = dd0 + td(days = n);
    return str(dt0.month)+"-"+str(dt0.day);
time0 = DT('2000');
time_start = int(time0/60);
#----------------------------------------------------------
def DTM(dtr):
    t = DT(dtr);
    if(t>time0): return t-time0;
    return t+(1440-time0);        
def DTL(dts):
    return [DTM(dts[0]),DTM(dts[1])];
#----------------------------------------------------------
#DATA STRIP
file = open('act.txt');
n = 0;
day,days = [],[];
for x in file:  
    if (n > 1):
        line = x.strip();
        naps = line.split('|');
        for i in range(0,len(naps)):
            dt = naps[i].split('-');
            day.append(DTL(dt));
        days.append(np.array(day));
        day.clear();
    n+=1
#----------------------------------------------------------
#RENDER
act = np.zeros((n-2,1440))
u = 0;
for day in days:
    for nap in day:
        for v in range(0,1440):
            if((v>=nap[0])&(v<nap[1])):
                act[u,v] = 1
        #print(nap)
    u+=1;
nOfDays = u;
state = np.ravel(np.copy(act))
for i in range(0,len(state)):
    if(state[i]>0):
        state[i]=-1
    else:
        state[i]=1
#----------------------------------------------------------
#BLUR
kernel = [[1,1,1],[1,5,1],[1,1,1]]
kernel /= np.linalg.norm(kernel)
act_k = c2d(act, kernel, mode='same')
ret, act_k = cv2.threshold(act_k,0.6, 255, cv2.THRESH_BINARY)
#----------------------------------------------------------
#DRAW
def DRW(act,fig,ax,cmap,title, tick_spacing_x,tick_spacing_y):
    ax.imshow(act, cmap)
    ax.autolimit_mode = 'round_numbers'
    ax.xmargin = 0
    ax.ymargin = 0
    ax.set_title('LD24 Sleep/Wake @ Sep-Nov2020' + title)
    ax.set_xlabel('Time')
    ax.set_xticklabels([x for x in np.arange(time_start-tick_spacing_x,24+tick_spacing_x,tick_spacing_x)]+[x for x in np.arange(tick_spacing_x,24,tick_spacing_x)])
    ax.xaxis.set_major_locator(ticker.MultipleLocator(60*tick_spacing_x))
    ax.set_ylabel('Day')
    ax.set_yticklabels([DD(int(x)) for x in np.arange(0-tick_spacing_y, (n-2),tick_spacing_y)])
    ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_y))
    ax.set_aspect(32)
    fig.savefig('LD24 SleepWake Sep-Nov2020' + title +".png",dpi=600)
    ax.set_aspect(32)
#----------------------------------------------------------
#VIEW
print("Number of Days: ",nOfDays)
fig, axs = plt.subplots(1,figsize=(20, 10))
DRW(act,fig,axs,'Greys',"",2,2);
#DRW(act_k,fig,axs[1],'Reds'," TRHD",2,4);
plt.tight_layout()
```

```python
class machine():
    pi = 3.14159265358    
    def __init__(self):
        self.D = -0.0353
        self.HA0 = 14.3
        self.LA0 = 2.4
        self.HA = 14.3
        self.LA = 2.4
        self.G = -0.3813
        self.BL = 12.2
        self.WC = -5.72
        self.WD = -1.51
        self.CM = 0
        self.CA = 2.5
        self.UA = .5
        self.UM = -.5
        self.P = 16.8
        self.A = 10.6
        self.B = -0.6
        #---------------Sleep Thresholds
        self.THR_H0 = 11.38
        self.THR_L0 = 8.38
        self.THR_H = 11.38
        self.THR_L = 8.38
        self.max_list_length = 20
        #--------------- Inital State @ 8PM - 2000
        self.TOD = 20
        self.state = 1
        self.oldState = 1
        self.bedtime = -2
        self.waketime = 8
        self.SS=4
        self.SW=13.3
        self.override = False
        self.overrideState = 1
        self.THR_H_LIST = [11.38]
        self.THR_L_LIST = [8.38]
        
    def setSleepState(self,stateIN):
        self.overrideState = stateIN
        self.override = True
    def state(self):
        return self.state
    def getTHRLists(self):
        return self.THR_H_LIST,self.THR_L_LIST
    def S(self):
        if(self.state>0):
            return self.LA + (self.SW-self.LA)*np.exp(self.D*self.TAW())
        else:
            return self.SPRIME()
    def SPRIME(self):
        BT = (self.BL-self.SS)/(self.G*(self.BL-self.HA))
        if((self.TOD-BT)>0):
            return self.S2(BT)
        return self.S1()
    def S1(self):
        return self.SS + self.TAS()*self.G*(self.BL-self.HA)
    def S2(self,BT):
        return self.HA + (self.BL-self.HA)*np.exp(self.G*(self.TAS()-BT))
    def W(self):
        return self.WC*np.exp(self.WD*self.TAW())
        
    def C(self):
        return self.CM+self.CA*np.cos((self.pi/12)*(self.TOD-self.P))
    def U(self):
        return self.UM+self.UA*np.cos((self.pi/6)*(self.TOD-self.P-3))
        
    def KSS(self):
        return self.A+self.B*(self.S()+self.C()+self.U()+self.W())
    
    def TRLD(self):
        return self.S()+self.C()+self.U()
    
    def toggleState(self):
        trld = self.TRLD()
        if(self.override):
            if(self.state!=self.overrideState):
                if(self.overrideState==1):
                    self.THR_H = trld
                    self.THR_H_LIST.append(trld)
                if(self.overrideState==-1):
                    self.THR_L = trld
                    self.THR_L_LIST.append(trld)
            self.state = self.overrideState
            self.override = False
            return;
        if(trld>=self.THR_H0):
            self.state = 1
        if(trld<=self.THR_L0):
            self.state = -1
    def TAW(self):
        return (self.TOD - self.waketime)
    def TAS(self):
        return (self.TOD - self.bedtime)
    def setTOD(self,TOD):
        self.TOD = TOD
    def adjustThresholds(self):
        if(len(self.THR_H_LIST)>self.max_list_length):
            self.THR_H_LIST.pop(0)
        if(len(self.THR_L_LIST)>self.max_list_length):
            self.THR_L_LIST.pop(0)
        self.THR_H0 = np.average(self.THR_H_LIST)
        self.THR_L0 = np.average(self.THR_L_LIST)
    def step(self,TOD):
        self.setTOD(TOD)
        Sstate = self.S()
        self.toggleState()
        if(self.state!=self.oldState):
            if(self.state==1):
                self.waketime = self.TOD
                self.SW = Sstate
            else:
                self.bedtime = self.TOD
                self.SS = Sstate
        self.oldState = self.state
        self.adjustThresholds()
        return self.KSS(),self.C()+self.U(),self.S(),self.TRLD(),self.state,self.THR_H0,self.THR_L0
```

```python
#----------------------------------------------------------
#DRAW
def DRW1D(fig,ax,title, tick_spacing_x,tick_spacing_y,time_st):
    ax.autolimit_mode = 'round_numbers'
    ax.xmargin = 0
    ax.ymargin = 0
    titleFull = 'LD24 Sleep/Wake @ Sep-Nov2020 = ' + title
    titleSave = 'LD24 SleepWake at Sep-Nov2020 -' + title
    ax.set_title(titleFull)
    ax.set_xlabel('Time')
    ax.set_xticklabels(([x for x in np.arange(time_st-3,24+tick_spacing_x,tick_spacing_x)]+[x for x in np.arange(tick_spacing_x,24+tick_spacing_x,tick_spacing_x)]+[x for x in np.arange(tick_spacing_x,24+tick_spacing_x,tick_spacing_x)]))
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_x))
    #ax.xaxis.set_major_locator(ticker.LinearLocator(27))#hacky solution - note incorrect (starts at 16 instead of 18!!) graph should begin at 8pm - 20
    ax.set_ylabel('KSS et al.')
    ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_y))
    first_legend = ax.legend(loc ='upper right') 
    ax.add_artist(first_legend) 
    fig.savefig(titleSave+".png",dpi=300)
def PLOT_AM(time,kss,cir,debt,trld,sleepstate,sleepstate0,thrh,thrl,time_st):
    #----------------------------------------------------------
    #VIEW
    fig, axs = plt.subplots(1,figsize=(20, 10))
    axs.plot(time,kss,label ="Karolinska Sleepiness Scale",color ="blue",linewidth = 6)
    axs.fill_between(time,[4]*len(time),kss, (kss>[4]*len(time)),label ="Alert Level",alpha = 0.2,facecolor ="navy",interpolate=True)
    axs.plot(time,cir,label ="Oscillator",color ="red")
    axs.plot(time,debt,label ="S Homeostasis",color ="cyan")
    axs.plot(time,trld,label ="Sleep Energy",color ="grey")
    axs.plot(time,thrh,label ="Average High Sleep Threshold",color ="grey",linestyle ='--')
    axs.plot(time,thrl,label ="Average Low Sleep Threshold",color ="grey",linestyle ='--')
    axs.plot(time,sleepstate,label ="State",color ="black")
    axs.plot(time,sleepstate0,label ="State0",color ="black",linestyle ='--')
    DRW1D(fig,axs,'AlertnessModel',2,1,time_st);
```

```python
time = []
kss = []
debt = []
cir = []
trld = []
thrh = []
thrl = []
sleepstate = []
sleepstate0 = []
startOffset = 50
days = 56 #add 1 to get first day from line # at act.txt
Ntime =24*days
minT = 60
Nstep = minT*Ntime
Sstep = minT*24*startOffset

m = machine()
for i in range (Sstep,Nstep):
    if(i<len(state)):
        m.setSleepState(state[i])
        sleepstate0.append(state[i])
    else:
        sleepstate0.append(0)
    timeN = 20+Ntime*((i-Sstep)/Nstep)
    k,c,s,t,stateR,th,tl = m.step(timeN)
    kss.append(k)
    cir.append(c)
    debt.append(s)
    trld.append(t)
    thrh.append(th)
    thrl.append(tl)
    sleepstate.append(stateR)
    time.append(timeN)
end = len(time)-1
st = end - minT*(48)+1
tst = int((time[end]-48)%24)
PLOT_AM(time[st:end],kss[st:end],cir[st:end],debt[st:end],trld[st:end],sleepstate[st:end],sleepstate0[st:end],thrh[st:end],thrl[st:end],tst)
```
